## Cell 4: Data Cleaning (Column Names)
#### This cell defines a function named clean_col_names to standardize the column names in the DataFrame. It replaces special characters and spaces with underscores to make them easier to work with.

```python
# --- 2. Data Cleaning ---

# Clean up column names
def clean_col_names(df):
    cols = df.columns
    new_cols = []
    for col in cols:
        new_col = col.replace('(UOM:Number), Scaling Factor:1', '')
        new_col = new_col.replace('(UOM:Ha(Hectare)), Scaling Factor:1', '')
        new_col = new_col.replace('Year (Jul - Jun)', 'Year')
        new_col = new_col.replace('-', '_')
        new_col = new_col.replace(' ', '_')
        new_col = new_col.replace('(', '')
        new_col = new_col.replace(')', '')
        new_col = new_col.replace(',', '')
        new_col = new_col.strip()
        new_cols.append(new_col)
    df.columns = new_cols
    return df
```

## Cell 5: Applying the Cleaning Function
#### This cell applies the clean_col_names function to the DataFrame df to update its column names.

```python
df = clean_col_names(df)
```

## Cell 6: Displaying Cleaned Column Names

## Univariate Analysis - Numerical Columns
#### This cell generates and saves histograms for all numerical columns in the DataFrame to visualize their distributions.

```python
# --- 3. Exploratory Data Analysis (EDA) ---

# Univariate Analysis

# Distribution of Numerical Columns
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

## Univariate Analysis - Categorical Columns
#### This cell creates and displays bar charts (count plots) for all categorical columns to show the frequency of each category.

```python
# Bar charts for categorical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()
for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.grid(True)
    plt.show() 
```

## Bivariate Analysis - Correlation Heatmap
#### This cell calculates the correlation matrix for all numerical columns and visualizes it as a heatmap, which helps in identifying relationships between variables.

```python
# Correlation Heatmap
plt.figure(figsize=(20, 15))
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

## Bivariate Analysis - Scatter Plot
#### This cell generates a scatter plot to visualize the relationship between two specific numerical columns: whollyownedholdingsnum and whollyownedareaha.

```python
# Scatter plot using the correct, cleaned column names
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, 
    x='whollyownedholdingsnum', 
    y='whollyownedareaha'       
)
plt.title('Wholly Owned Holdings vs. Area')
plt.xlabel('Number of Wholly Owned Holdings')
plt.ylabel('Area of Wholly Owned Holdings (Ha)')
plt.grid(True)
plt.show() 
```

## Bivariate Analysis - Box Plot
#### This cell creates a box plot to compare the distribution of the whollyownedareaha column across the different categories in the categoryholdings column.

```python
# Box plot using the correct, cleaned column names
plt.figure(figsize=(12, 8))
sns.boxplot(
    data=df, 
    x='categoryholdings',     
    y='whollyownedareaha'     
)
plt.title('Area of Wholly Owned Holdings by Category')
plt.xlabel('Category of Holdings')
plt.ylabel('Area of Wholly Owned Holdings (Ha)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
