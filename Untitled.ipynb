## Step 1: Import Required Libraries
#### This cell imports essential Python libraries (pandas, numpy, matplotlib, and seaborn) used for data handling and visualization, and sets default styling for plots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Show plots inside Jupyter notebooks
%matplotlib inline

# Set plot style
sns.set(style="whitegrid")
```

## Step 2: Load the Dataset
#### Loads the CSV file into a DataFrame named df and displays the first five rows to get an initial view of the data structure.

```python
# Load your dataset
df = pd.read_csv('gurupreetdhande20_17538577291145594.csv')

# Preview the first 5 rows
df.head()
```

## Dataset Overview
#### Prints the shape of the dataset, data types, non-null counts, and basic statistics (mean, count, etc.) to understand the structure and contents of the dataset.

```python
# Shape of dataset
print("Shape of the dataset:", df.shape)

# Info about data types and non-null counts
df.info()

# Summary statistics (numeric + categorical)
df.describe(include='all')
```

## Step 4: Check Missing Values
#### Identifies columns with missing data numerically and visually highlights them using a heatmap, which helps decide whether imputation or dropping is needed.

```python
# Total missing values
print("Missing values in each column:\n", df.isnull().sum())

# Visualize missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Value Heatmap")
plt.show()
```

## Step 5: Data Types and Column Categories
#### Separates the features into categorical and numerical columns for easier analysis in later steps and prints the names of each.

```python
# Separate categorical and numerical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

print("Categorical columns:\n", categorical_cols)
print("Numerical columns:\n", numerical_cols)
```

## Step 6: Univariate Analysis (Categorical Columns)
#### Generates count plots for each categorical column to visualize the frequency distribution of each category, helping detect imbalance or dominance.

```python
# Count plots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f"Count Plot: {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Step 7: Univariate Analysis (Numerical Columns)
#### Creates histograms for all numerical columns to inspect their distributions and detect skewness, gaps, or outliers.

```python
# Histograms of numerical columns
df[numerical_cols].hist(figsize=(18, 15), bins=30, color='skyblue')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

## Step 8: Boxplots to Detect Outliers
#### Draws boxplots for each numerical feature to visually identify outliers and understand value ranges.

```python
# Boxplots for each numerical feature
for col in numerical_cols:
    plt.figure(figsize=(10, 2))
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot: {col}")
    plt.tight_layout()
    plt.show()
```

## Step 9: Correlation Matrix (Numerical Features)
#### Computes and plots a heatmap of correlation between numerical features to detect potential multicollinearity or linear relationships.

```python
# Compute correlation matrix
corr_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

## Step 10: Pairplot (Optional: Small Sample)
#### Generates pairwise scatter plots for a subset of numerical features using a random 500-row sample to explore variable relationships in more detail.

```python
# Optional: Use only if few features or rows
sample_df = df.sample(500)  # sample for faster rendering
sns.pairplot(sample_df[numerical_cols[:5]])  # adjust for fewer features
plt.suptitle("Pairplot of Sample Data", y=1.02)
plt.show()
```
